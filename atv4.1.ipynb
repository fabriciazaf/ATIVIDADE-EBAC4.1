## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

# Define uma semente para que os números aleatórios sejam reproduzíveis
np.random.seed(42)

# Criando DataFrames de acertos para cada assunto, conforme especificado
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])
df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])
df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])
df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])
df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])
df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])


#01 Cálculo do percentual de acerto para cada assunto com arredondamento e formatação
df_mat['Percentual'] = ((df_mat['Qt_acertos'] / 24) * 100).round(2).astype(str) + '%'
df_por['Percentual'] = ((df_por['Qt_acertos'] / 18) * 100).round(2).astype(str) + '%'
df_geo['Percentual'] = ((df_geo['Qt_acertos'] / 8) * 100).round(2).astype(str) + '%'
df_ing['Percentual'] = ((df_ing['Qt_acertos'] / 8) * 100).round(2).astype(str) + '%'
df_his['Percentual'] = ((df_his['Qt_acertos'] / 8) * 100).round(2).astype(str) + '%'
df_fis['Percentual'] = ((df_fis['Qt_acertos'] / 12) * 100).round(2).astype(str) + '%'
df_qui['Percentual'] = ((df_qui['Qt_acertos'] / 12) * 100).round(2).astype(str) + '%'

# Exibindo os primeiros 5 resultados de cada DataFrame para verificação
print("Percentual de acertos em Matemática:\n", df_mat.head())
print("Percentual de acertos em Português:\n", df_por.head())
print("Percentual de acertos em Geografia:\n", df_geo.head())
print("Percentual de acertos em Inglês:\n", df_ing.head())
print("Percentual de acertos em História:\n", df_his.head())
print("Percentual de acertos em Física:\n", df_fis.head())
print("Percentual de acertos em Química:\n", df_qui.head())
````

````python
#02 - (operações entre *DataFrames) Calcule o total de acertos de cada aluno.

import pandas as pd
import numpy as np

# Define uma semente para que os números aleatórios sejam reproduzíveis
np.random.seed(42)

# Criando DataFrames de acertos para cada assunto, conforme especificado
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Matemática'])
df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Português'])
df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Geografia'])
df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Inglês'])
df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['História'])
df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Física'])
df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Química'])

# Concatenando todos os DataFrames horizontalmente
df_acertos = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)

# Calculando o total de acertos por aluno (somando as linhas)
df_acertos['Total_acertos'] = df_acertos.sum(axis=1)

# Exibir os primeiros 5 resultados
print("Total de acertos de cada aluno:\n", df_acertos.head())


````

````python
#03 - calcule o percentual de acertos de cada aluno

import pandas as pd
import numpy as np

# Define uma semente para que os números aleatórios sejam reproduzíveis
np.random.seed(42)

# Criando DataFrames de acertos para cada assunto, conforme especificado
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Matemática'])
df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Português'])
df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Geografia'])
df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Inglês'])
df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['História'])
df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Física'])
df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Química'])

# Concatenando todos os DataFrames horizontalmente
df_acertos = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)

# Calculando o total de acertos por aluno
df_acertos['Total_acertos'] = df_acertos.sum(axis=1)

# Definindo o total de questões
total_questoes = 24 + 18 + 8 + 8 + 8 + 12 + 12

# Calculando o percentual de acertos por aluno e arredondando para duas casas decimais
df_acertos['Percentual_acertos'] = ((df_acertos['Total_acertos'] / total_questoes) * 100).round(2).astype(str) + '%'

# Exibindo os primeiros 5 resultados para verificação
print("Percentual de acertos de cada aluno:\n", df_acertos[['Total_acertos', 'Percentual_acertos']].head())
````

````python
#04 - Suponha que a nota de corte para a segunda fase seja 45. 

#filtrando alunos com nota maior que 45
alunos_acima_corte = df_acertos[df_acertos['Total_acertos'] > 45]

#contando alunos nota >45
num_alunos_acima_corte = alunos_acima_corte.shape[0]

print("Número de alunos com nota maior que 45:", num_alunos_acima_corte)
````

2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python

#01 -  Identificando os alunos desqualificados (aqueles que zeraram em Matemática, Física ou Química)
df_desqualificados = df_acertos[(df_acertos['Matemática'] == 0) | 
                                (df_acertos['Física'] == 0) | 
                                (df_acertos['Química'] == 0)]


# 02 - Contando o número de alunos desqualificados
num_desqualificados = df_desqualificados.shape[0]

# 03 - Calculando a média de acertos em História e Geografia para os alunos desqualificados
media_hist_geo_desqualificados = df_desqualificados[['História', 'Geografia']].mean()

# Identificando os alunos que passaram para a segunda fase:
# - Total_acertos > 45 - Não zeraram em Matemática, Física ou Química
df_aprovados = df_acertos[(df_acertos['Total_acertos'] > 45) & 
                          (df_acertos['Matemática'] > 0) & 
                          (df_acertos['Física'] > 0) & 
                          (df_acertos['Química'] > 0)]

# Resultados
print("Número de alunos desqualificados:", num_desqualificados)
print("Média de acertos em História e Geografia dos alunos desqualificados:\n", media_hist_geo_desqualificados)
print("Número de alunos que passaram para a segunda fase:", df_aprovados.shape[0])
````

## 4) Vacinação II
Gere um *DataFrame* que contenha somente os estabelecimentos que aplicaram vcinas a menores de 18 anos. Nesse *DataFrame* devem conter somente os dados dos estabelecimentos, mais uma coluna sendo a quantidade de vacinas que o estabelecimento aplicou a menores de 18 anos.  
  
1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**  
3. Nesse *DataFrame* faça uma contagem do campo ```vacina_categoria_nome```.
3. a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.  
4. usando o método *.loc*, selecione somente essas variáveis  
5. Aplique o método **.drop_duplicates** e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento  

````python
# 1. Criar uma cópia do DataFrame original com vacinas aplicadas a menores de 18 anos
df_menores_18 = df[df['paciente_idade'] < 18].copy()

# 2. Criar uma lista das colunas do DataFrame
colunas_menores_18 = df_menores_18.columns.tolist()
print(f"Colunas do DataFrame: {colunas_menores_18}")

# 3. Contagem do campo vacina_categoria_nome
contagem_categoria = df_menores_18['vacina_categoria_nome'].value_counts()
print("Contagem de categorias de vacinas:")
print(contagem_categoria)

# 4. Escolher colunas referentes ao estabelecimento
colunas_estabelecimento = [
    'estabelecimento_valor',
    'estabelecimento_razaoSocial',
   'estalecimento_noFantasia', 
    'estabelecimento_municipio_codigo',
    'estabelecimento_municipio_nome',
    'estabelecimento_uf'
]

# 5. Selecionar somente as variáveis referentes ao estabelecimento
df_estabelecimentos = df_menores_18.loc[:, colunas_estabelecimento]

# 6. Remover duplicados e criar uma lista única de estabelecimentos
estabelecimentos_unicos = df_estabelecimentos.drop_duplicates()

# Contar o número de vacinas aplicadas por estabelecimento
vacinas_por_estabelecimento = df_menores_18.groupby('estabelecimento_municipio_codigo').size().reset_index(name='quantidade_vacinas')

# Mesclar com os dados únicos dos estabelecimentos
df_resultado = estabelecimentos_unicos.merge(vacinas_por_estabelecimento, on='estabelecimento_municipio_codigo')

#wxibindo em forma de tabela
df_resultado


````
